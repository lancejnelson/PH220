## A simple example.

Let's see if we can generate a contour plot for the electric potential due to just a single point charge, located at $\vec{r} = (1,1)$ m.  



```python
from numpy import arange,meshgrid,sqrt,array
from numpy.linalg import norm
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D

x = arange(-2,2,.01)
y = arange(-2,2,.01)
X,Y = meshgrid(x,y)

k = 8.99e9
q = 100e-9

rCharge = array([1,1])

V = k * q/sqrt((X-rCharge[0])**2 + (Y-rCharge[1])**2)
fig,ax = pyplot.subplots(figsize = (10,10))
contourLocations = [100,250,500,1000,2000,5000]
CP = ax.contour(X,Y,V,contourLocations)  
ax.clabel(CP, inline=True, fontsize=10)
pyplot.show()
```
