## Overview

Previously, you learned how to generate vector field plots when we were learning about the electric field.  Let's use those same skills to generate vector field plots of magnetic fields.

Below you will find the code to plot the magnetic field due to a current carrying wire. The direction of the current is into the screen. 



```python



from numpy import linspace,meshgrid,sqrt,pi,log
from numpy.linalg import norm
from matplotlib import pyplot

x = linspace(-6,15,20)
y = linspace(-6,6,20)

X,Y = meshgrid(x,y)

I = 5
mu = 1.26e-6

R = sqrt(X**2 + Y**2)
Bx = mu * I/2/pi/R * Y/R
By = mu * I/2/pi/R * -X/R

vecNorms = norm([Bx,By],axis = 0)  # get the magnitude of every e-field vector


fig,ax = pyplot.subplots(figsize=(10,10))
vectorPlot=ax.quiver(X,Y,Bx/vecNorms,By/vecNorms,log(vecNorms*10**8),cmap='gist_rainbow')
ax.set_aspect(1)
pyplot.colorbar(vectorPlot,orientation="horizontal")
pyplot.show()
```
