## Problem Statement

The code in the cell below is a solution to the following problem:

A straight rod of length $L = 0.5$ m has $Q = 5.0$ nC spread uniformly across it.
Plot the electric field created by this charged rod along a line that bisects it.

```python
from numpy import linspace,sqrt
from matplotlib import pyplot

k = 8.99e9
q = 5e-9
l = 0.5
r = linspace(0.1,10,1000)
E = k * q/(r * sqrt(r**2 + l**2/4))

pyplot.figure(figsize = (10,10))
pyplot.plot(r,E)
pyplot.ylim(0,100)

pyplot.show()



```
