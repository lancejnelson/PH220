## Example

Below you will find the python code that plots the electric potential due to three point charges in the x-y plane.  By experimenting with the code learn what each line does and add a comment to explain.  Note: Adding print statements and looking at the output is a good way to figure out what is going on.


If you'd like more reading material on how to generate advanced plots, you can refer to chapter 9 in [the python book](https://byui-physics.github.io/pythonbook/).

```python
from numpy import arange,meshgrid,sqrt
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D

x = arange(-2,2,.012)
y = arange(-2,2,.01)
X,Y = meshgrid(x,y)

k = 8.99e9
q1 = -100e-9
q2 = -100e-9
q3 = 200e-9

V = k * q1/sqrt((X-0.5)**2 + Y**2) + k * q2/sqrt((X+0.5)**2 + Y**2) + k * q3/sqrt((X-0.5)**2 + (Y-0.5)**2)
fig = pyplot.figure(figsize = (75,75))
ax = fig.gca(projection='3d')
ax.plot_wireframe(X,Y,V)  # you can also try plot_surface for a little different look.
ax.set_zlim(-25000,25000)
pyplot.show()
```
